## Exploratory data Analysis - Terrorism
### Problem Statement : Find out the security issues and hot zone of terrorism.
#### Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
Terror = pd.read_csv('globalterrorismdb_0718dist.csv', encoding = 'ISO-8859-1', low_memory = False) 
Terror.head()
```

```python
Terror.shape
```

```python
Terror.columns
```

```python
Terror.describe()
```

```python
Terror.drop_duplicates()
```

```python
Terror = Terror[['iyear', 'imonth', 'iday', 'country_txt', 'success', 'weapdetail', 'region_txt', 'attacktype1_txt', 'targtype1_txt', 'gname', 'weaptype1_txt', 'latitude', 'nkill', 'city' , 'longitude']]
Terror
```

```python
Terror.isnull().sum()
```

```python
Terror['weapdetail'].fillna('Unknown', inplace = True)
Terror['city'].fillna('Unknown', inplace = True)
```

```python
#dropping Rows with missing values
Terror.dropna(axis = 0, how = 'any')
Terror.dropna(axis = 1, how = 'any')
Terror
```

#### Visualizing the data

#### Year Wise

```python
plt.subplots(figsize = (16,5))
sns.countplot(x = 'iyear', data = Terror)
plt.title('****Attacks Each Year****', fontsize = 22)
plt.ylabel('Attack Count',fontsize = 22)
plt.xlabel('Year',fontsize = 22)
plt.xticks(rotation = 45)
plt.show()
```

#### Success and Failure Wise

```python
plt.subplots(figsize = (16,5))
sns.countplot(x = 'iyear', hue = 'success', data = Terror)
plt.title('****Success and Failure of Attacks****', fontsize = 22)
plt.ylabel('Attack Count',fontsize = 22)
plt.xlabel('Year',fontsize = 22)
plt.xticks(rotation = 45)
plt.show()
```

#### Region Wise

```python
plt.subplots(figsize = (10,5))
sns.countplot(y = 'region_txt', order = Terror['region_txt'].value_counts().index, data = Terror)
plt.title('****Region Wise Terrorist Attacks****', fontsize = 22)
plt.xlabel('Attack Count',fontsize = 22)
plt.ylabel('Regions',fontsize = 22)
plt.show()
```

#### Country Wise(Top 20)

```python
plt.subplots(figsize = (10,5))
sns.countplot(y = 'country_txt', order = Terror['country_txt'].value_counts()[:20].index, data = Terror)
plt.title('****Country wise Terroist Attacks****', fontsize = 22)
plt.xlabel('Attack Count',fontsize = 22)
plt.ylabel('Countries',fontsize = 22)
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (12,5))
Terror.groupby(['country_txt'])['nkill'].sum().sort_values(ascending = False)[:20].plot.bar(color = 'pink')
Terror['country_txt'].value_counts().sort_values(ascending = False)[:20].plot.bar()
l=plt.legend()
l.get_texts()[1].set_text('Number of Attacks')
plt.xlabel('countries', fontsize = 22)
plt.title('****Country Wise Terrorist Attacks and Total Deaths****', fontsize = 22)

plt.show()
```

#### Citiy Wise(Top 20)

```python
plt.subplots(figsize = (10,5))
sns.countplot(y = 'city', order = Terror['city'].value_counts()[:20].index, data = Terror)
plt.title('****City Wise Terrorist Attacks****', fontsize = 22)
plt.xlabel('Attack Count',fontsize = 22)
plt.ylabel('Cities',fontsize = 22)
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (12,5))
Terror.groupby(['city'])['nkill'].sum().sort_values(ascending = False)[:20].plot.bar(color = 'pink')
Terror['city'].value_counts().sort_values(ascending = False)[:20].plot.bar()
l=plt.legend()
l.get_texts()[1].set_text('Number of Attacks')
plt.xlabel('cities', fontsize = 22)
plt.title('****City Wise Terrorist Attacks and Total Deaths****', fontsize = 22)

plt.show()
```

#### Type Wise

```python
plt.subplots(figsize = (10,5))
sns.countplot(y = 'attacktype1_txt', order = Terror['attacktype1_txt'].value_counts().index, data = Terror)
plt.title('****Types wise Terrorist Attacks****', fontsize = 22)
plt.xlabel('Attack Count',fontsize = 22)
plt.ylabel('Attack Types',fontsize = 22)
plt.xticks(rotation = 0)
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (12,5))
Terror.groupby(['attacktype1_txt'])['nkill'].sum().sort_values(ascending = False)[:20].plot.bar(color = 'pink')
Terror['attacktype1_txt'].value_counts().sort_values(ascending = False)[:20].plot.bar()
l=plt.legend()
l.get_texts()[1].set_text('Number of Attacks')
plt.xlabel('Attack Type', fontsize = 22)
plt.title('****Attack Types in Terrorist Attacks and Total Deaths****', fontsize = 22)

plt.show()
```

```python
plt.subplots(figsize = (10,6))
sns.countplot(y = 'targtype1_txt', order = Terror['targtype1_txt'].value_counts().index, data = Terror)
plt.title('****Targets wise  Terrorist Attacks****', fontsize = 22)
plt.xlabel('Attack Count',fontsize = 22)
plt.ylabel('Target Types',fontsize = 22)
plt.show()
```
